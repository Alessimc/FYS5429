# Combined plots

```python
plt.rcParams.update({'font.size': plt.rcParams['font.size'] + 2})
```

All models loose skill beteen may and august, with june being the most extreme. This coincides with rapid melting due to 24-h sunlight. During this period, melt ponds are created on the sea ice surface and the sea ice dielectric properties themselves change

```python
# Turbo colormap from matplotlib
cmap = plt.cm.turbo

# Define the range for yellow to red in Turbo (approx 0.6 to 1.0)
n_colors = 20
start, end = 0.0, 0.7
values = np.linspace(start, end, n_colors)

# Get RGB colors
colors = [cmap(v) for v in values]
hex_colors = [mcolors.to_hex(c) for c in colors]
```

```python


fig, ax = plt.subplots()

line_w = 1.8

# Create custom legend entries
custom_lines = [
    Line2D([0], [0], color='gray', lw=line_w, label='V - Pol'),
    Line2D([0], [0], color='gray', lw=line_w, linestyle='--', label='H - Pol'),
]

start_col = 0
step = 3

custom_lines2 = [
    Line2D([0], [0], color=hex_colors[start_col], lw=line_w, label='REDNet'),
    Line2D([0], [0], color=hex_colors[start_col + step], lw=line_w, label='VAE'),
    Line2D([0], [0], color=hex_colors[start_col + 2*step], lw=line_w, label='ELAN'),
    Line2D([0], [0], color=hex_colors[start_col + 3*step], lw=line_w, label='SwinIR'),
    Line2D([0], [0], color=hex_colors[start_col + 4*step], lw=line_w, label='SRCNN'),
]

# Plot data
ax.set_facecolor("gainsboro")  


ax.plot(REDNet256["date"], (REDNet256["psnr_value_v"] - REDNet256["psnr_baseline_v"]).rolling(window=10, center=True).mean(), label="REDNet V-Pol", lw=line_w, c=hex_colors[start_col])
ax.plot(REDNet256["date"], (REDNet256["psnr_value_h"] - REDNet256["psnr_baseline_h"]).rolling(window=10, center=True).mean(), label="REDNet H-Pol", lw=line_w, c=hex_colors[start_col], ls="--")

ax.plot(REDNetVAE["date"], (REDNetVAE["psnr_value_v"] - REDNetVAE["psnr_baseline_v"]).rolling(window=10, center=True).mean(), label="VAE V-Pol", lw=line_w, c=hex_colors[start_col + step])
ax.plot(REDNetVAE["date"], (REDNetVAE["psnr_value_h"] - REDNetVAE["psnr_baseline_h"]).rolling(window=10, center=True).mean(), label="VAE H-Pol", lw=line_w, c=hex_colors[start_col + step], ls="--")

ax.plot(ELAN["date"], (ELAN["psnr_value_v"] - ELAN["psnr_baseline_v"]).rolling(window=10, center=True).mean(), label="ELAN V-Pol", lw=line_w, c=hex_colors[start_col + 2*step])
ax.plot(ELAN["date"], (ELAN["psnr_value_h"] - ELAN["psnr_baseline_h"]).rolling(window=10, center=True).mean(), label="ELAN H-Pol", lw=line_w, c=hex_colors[start_col + 2*step], ls="--")

ax.plot(SwinIR["date"], (SwinIR["psnr_value_v"] - SwinIR["psnr_baseline_v"]).rolling(window=10, center=True).mean(), label="SwinIR V-Pol", lw=line_w, c=hex_colors[start_col + 3*step])
ax.plot(SwinIR["date"], (SwinIR["psnr_value_h"] - SwinIR["psnr_baseline_h"]).rolling(window=10, center=True).mean(), label="SwinIR H-Pol", lw=line_w, c=hex_colors[start_col + 3*step], ls="--")

ax.plot(SRCNN_W256["date"], (SRCNN_W256["psnr_value_v"] - SRCNN_W256["psnr_baseline_v"]).rolling(window=10, center=True).mean(), label="SRCNN V-Pol", lw=line_w, c=hex_colors[start_col + 4*step])
ax.plot(SRCNN_W256["date"], (SRCNN_W256["psnr_value_h"] - SRCNN_W256["psnr_baseline_h"]).rolling(window=10, center=True).mean(), label="SRCNN H-Pol", lw=line_w, c=hex_colors[start_col + 4*step], ls="--")

# Format x-axis to show only months
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))

# Labels and layout
ax.set_ylabel("PSNR improvement")
plt.xticks(rotation=45)
plt.tight_layout()

# Add both legends
legend1 = ax.legend(handles=custom_lines, loc="lower right", framealpha=0.7)
ax.add_artist(legend1)  # Add first legend manually
ax.legend(handles=custom_lines2, loc="lower left", framealpha=0.7)

# # Hide unwanted spines

for side in ['top', 'right']:
    ax.spines[side].set_visible(False)

plt.savefig("../final_figures/combined_psnr_timeserie.pdf", bbox_inches="tight")
plt.show()
```
