```python
import pandas as pd 
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': plt.rcParams['font.size'] + 2})
```

```python
REDNet10_masked = pd.read_csv('REDNet10_24h_batchsize8_samplesall_normalized_masked_loss.csv')
REDNet20_masked = pd.read_csv('REDNet20_24h_batchsize8_samplesall_normalized_masked_loss.csv')
REDNet30_masked = pd.read_csv('REDNet30_24h_batchsize8_samplesall_normalized_masked_loss.csv')

REDNet10 = pd.read_csv('REDNet10_24h_batchsize8_samplesall_normalized.csv')
REDNet20 = pd.read_csv('REDNet20_24h_batchsize8_samplesall_normalized.csv')
REDNet30 = pd.read_csv('REDNet30_24h_batchsize8_samplesall_normalized.csv')
REDNet256 = pd.read_csv('REDNet256_24h_batchsize8_samplesall_normalized.csv')
```

```python
plt.plot(REDNet10_masked['epoch'], REDNet10_masked['val_loss'], label="REDNet10_masked_loss")
plt.plot(REDNet20_masked['epoch'], REDNet20_masked['val_loss'], label="REDNet20_masked_loss")
plt.plot(REDNet30_masked['epoch'], REDNet30_masked['val_loss'], label="REDNet30_masked_loss")

plt.ylim(00.00165,0.0035)
plt.xlim(1,45)
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.legend()
# plt.savefig("../autoencoder_figures/autencoder_masked_val_loss.png", bbox_inches="tight")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6.4, 3.2))


ax.plot(REDNet10['epoch'], REDNet10['val_loss'], label="REDNet10")
ax.plot(REDNet20['epoch'], REDNet20['val_loss'], label="REDNet20")
ax.plot(REDNet30['epoch'], REDNet30['val_loss'], label="REDNet30")
ax.plot(REDNet256['epoch'], REDNet256['val_loss'], label="REDNet256")

ax.set_facecolor("gainsboro")  


plt.ylim(0.56335,0.5640)
plt.xlim(1,83)
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.legend(framealpha=0.7)

for side in ['top', 'right']:
    ax.spines[side].set_visible(False)

# plt.savefig("../autoencoder_figures/autencoder_val_loss.png", bbox_inches="tight")
plt.savefig("../../final_figures/autencoder_val_loss.pdf", bbox_inches="tight")
plt.show()
```
