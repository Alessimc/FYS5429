```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
REDNet10 = pd.read_csv("REDNet10_24h_batchsize8_samplesall_normalized.csv")
REDNet10_masked_loss = pd.read_csv("REDNet10_24h_batchsize8_samplesall_normalized_masked_loss.csv")

REDNet20 = pd.read_csv("REDNet20_24h_batchsize8_samplesall_normalized.csv")
REDNet20_masked_loss = pd.read_csv("REDNet20_24h_batchsize8_samplesall_normalized_masked_loss.csv")

REDNet30 = pd.read_csv("REDNet30_24h_batchsize8_samplesall_normalized.csv")
REDNet30_masked_loss = pd.read_csv("REDNet30_24h_batchsize8_samplesall_normalized_masked_loss.csv")
```

```python
plt.plot(REDNet10['epoch'], REDNet10['val_loss'], label="val")
plt.plot(REDNet10['epoch'], REDNet10['train_loss'], label = "train")
plt.legend()
plt.show()
```

```python
plt.plot(REDNet10_masked_loss['epoch'], REDNet10_masked_loss['val_loss'], label="val masked loss")
plt.plot(REDNet10_masked_loss['epoch'], REDNet10_masked_loss['train_loss'], label = "train  masked loss")
plt.legend()
plt.show()
```

```python
plt.plot(REDNet10['epoch'], REDNet10['val_loss'], label="REDNet10 val")
plt.plot(REDNet20['epoch'], REDNet20['val_loss'], label="REDNet20 val")
plt.plot(REDNet30['epoch'], REDNet30['val_loss'], label="REDNet30 val")
plt.legend()
plt.show()
```

```python
plt.plot(REDNet10_masked_loss['epoch'], REDNet10_masked_loss['val_loss'], label="REDNet10 val masked loss")
plt.plot(REDNet20_masked_loss['epoch'], REDNet20_masked_loss['val_loss'], label="REDNet20 val masked loss")
plt.plot(REDNet30_masked_loss['epoch'], REDNet30_masked_loss['val_loss'], label="REDNet30 val masked loss")
plt.legend()
plt.show()
```
