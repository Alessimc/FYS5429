```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
SRCNN1_batchsize8 = pd.read_csv("srcnn_model1_24h_batchsize8_samplesall_normalized.csv")
SRCNN1_batchsize16 = pd.read_csv("srcnn_model1_24h_batchsize16_samplesall_normalized.csv")
SRCNN1_batchsize32 = pd.read_csv("srcnn_model1_24h_batchsize32_samplesall_normalized.csv")

SRCNN2_batchsize8 = pd.read_csv("srcnn_model2_24h_batchsize8_samplesall_normalized.csv")
SRCNN2_batchsize16 = pd.read_csv("srcnn_model2_24h_batchsize16_samplesall_normalized.csv")
SRCNN2_batchsize64 = pd.read_csv("srcnn_model2_24h_batchsize64_samplesall_normalized.csv")

SRCNN3_batchsize8 = pd.read_csv("srcnn_model3_24h_batchsize8_samplesall_normalized.csv")
SRCNN4_batchsize8 = pd.read_csv("srcnn_model4_24h_batchsize8_samplesall_normalized.csv")
SRCNN5_batchsize8 = pd.read_csv("srcnn_model5_24h_batchsize8_samplesall_normalized.csv")
SRCNN6_batchsize8 = pd.read_csv("srcnn_model6_24h_batchsize8_samplesall_normalized.csv")
```

```python
plt.plot(SRCNN1_batchsize8['epoch'], SRCNN1_batchsize8['val_loss'], label="batch size 8")
plt.plot(SRCNN1_batchsize16['epoch'], SRCNN1_batchsize16['val_loss'], label="batch size 16")
plt.plot(SRCNN1_batchsize32['epoch'], SRCNN1_batchsize32['val_loss'], label="batch size 32")
plt.ylim(00.002,0.004)
plt.legend()
```

```python
plt.plot(SRCNN2_batchsize8['epoch'], SRCNN2_batchsize8['val_loss'], label="batch size 8")
plt.plot(SRCNN2_batchsize16['epoch'], SRCNN2_batchsize16['val_loss'], label="batch size 16")
plt.plot(SRCNN2_batchsize64['epoch'], SRCNN2_batchsize64['val_loss'], label="batch size 64")
plt.ylim(00.001,0.004)
plt.legend()
```

```python
plt.plot(SRCNN1_batchsize8['epoch'], SRCNN1_batchsize8['val_loss'], label="SRCNN1")
plt.plot(SRCNN2_batchsize8['epoch'], SRCNN2_batchsize8['val_loss'], label="SRCNN2") # wider
plt.plot(SRCNN3_batchsize8['epoch'], SRCNN3_batchsize8['val_loss'], label="SRCNN3") # deeper
# plt.plot(SRCNN4_batchsize8['epoch'], SRCNN4_batchsize8['val_loss'], label="SRCNN4") #wider and deeper
# plt.plot(SRCNN5_batchsize8['epoch'], SRCNN5_batchsize8['val_loss'], label="SRCNN5") # even wider and deeper
plt.plot(SRCNN6_batchsize8['epoch'], SRCNN6_batchsize8['val_loss'], label="SRCNN6")

plt.legend()
plt.ylim(00.0015,0.004)
plt.show()
```
